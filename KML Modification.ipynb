```python
##Loading libararies
import xml.etree.ElementTree as ET
```

```python
### Registering namespaces so we don't get undesired special characters
ET.register_namespace('', "http://www.opengis.net/kml/2.2")
ET.register_namespace('gx', "http://www.google.com/kml/ext/2.2")
ET.register_namespace('atom', "http://www.w3.org/2005/Atom")
ET.register_namespace('xal', "urn:oasis:names:tc:ciq:xsdschema:xAL:2.0")
```

```python
namespaces = {'kml': 'http://www.opengis.net/kml/2.2', } # add more as needed

# load data from KML file
tree = ET.parse('Zwischenlager_Brandenburg.kml')
root = tree.getroot()

##Tags which we want to search in the KML file
xpath = ".//kml:Placemark/kml:Point/kml:coordinates"

###editing the KML file
##Correcting the co-ordinates which were flipped
for elem in root.findall(xpath, namespaces):
    elem.text = ' '.join(elem.text.split(',')[::-1])
    print("coordinates=", ' '.join(elem.text.split(',')[::-1]))
    
#Writnng the output to a kml file
tree.write('output.kml')
```
